### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
def find_consistent_cars(data):
    """
    Find car models with lowest variation in MPG across multiple years.
    
    Parameters:
    data: DataFrame with columns 'car_name', 'model_year', 'mpg'
    
    Returns:
    DataFrame with consistent cars showing appearances and average MPG
    """
    cars_multiple_years = data.groupby('name').filter(lambda x: x['model_year'].nunique() > 1)
    
    mpg_stats = cars_multiple_years.groupby('name').agg({
        'mpg': ['std', 'mean'],
        'model_year': 'nunique'
    }).round(2)
    
    
    mpg_stats.columns = ['std_mpg', 'avg_mpg', 'appearances']
    
    
    consistent_cars = mpg_stats[
        (mpg_stats['std_mpg'] < 1.0) & 
        (mpg_stats['appearances'] >= 2)
    ]
    
    
    if len(consistent_cars) > 0:
        min_std = consistent_cars['std_mpg'].min()
        lowest_variation_cars = consistent_cars[consistent_cars['std_mpg'] == min_std]
        
        
        result = lowest_variation_cars.sort_values(
            by=['appearances', 'avg_mpg'], 
            ascending=[False, False]
        )
        
        return result[['appearances', 'avg_mpg']]
    else:
        return pd.DataFrame(columns=['appearances', 'avg_mpg'])
    
print(data)
print("\nConsistent cars with lowest MPG variation:")
result = find_consistent_cars(data)
print(result)
```
